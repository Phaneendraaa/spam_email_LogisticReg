```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"C:\Users\nagap\Downloads\mail_data.csv")
```

```python
df.head()
```

```python
df = df.where((pd.notnull(df)),'')
```

```python
df.head()
```

```python
df["Category"] = df["Category"].str.strip()
```

```python
df["Category"] = df["Category"].replace("ham",1)
df["Category"] = df["Category"].replace("spam",0)
```

```python
df["Category"].value_counts()
```

```python
X = df["Message"]
y = df["Category"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
features = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
```

```python
X_train_featured = features.fit_transform(X_train)
X_test_featured = features.transform(X_test)
```

```python
spam_model = LogisticRegression()
```

```python
spam_model.fit(X_train_featured,y_train)
```

```python
train_predictions = spam_model.predict(X_train_featured)
train_accuracy = accuracy_score(y_train,train_predictions)
train_accuracy
```

```python
test_predictions = spam_model.predict(X_test_featured)
test_accuracy = accuracy_score(y_test,test_predictions)
test_accuracy
```
